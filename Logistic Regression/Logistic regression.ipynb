```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( "diabetes.csv" )
X = df.iloc[:,:1].values
Y = df.iloc[:,-1:].values
```

```python
c=0
m=0
for i in range(1000):
    z=1/(1 + np.exp(-(c+m*X)))
    dm=sum((z-Y)*X)
    dc=sum(z-Y)
    c=c-(dc*0.001)
    m=m-(dm*0.001)
print("m=",m,"c=",c)
```

```python
Y_Pred=1/(1 + np.exp(-(c+m*X)))
Y_Pred
```

```python
z=np.where( Y_Pred > 0.5, 1, 0 ) 
z
```

```python
correctly_classified = 0    


# counter    
count = 0    
for count in range( np.size( z ) ) :  

    if z[count] == Y[count] :            
        correctly_classified = correctly_classified + 1

    count = count + 1

print( "Accuracy on test set by our model       :  ", ( 
  correctly_classified / count ) * 100 )
```

```python
plt.scatter(X,z)
plt.plot(X,Y_Pred)
plt.show()
```

### logistic regression with multivariable

```python
df1=pd.read_csv("pic.csv")
dum=pd.get_dummies(df1,columns=['det'])
df1=pd.melt(dum,id_vars=['variance','maximum'],
       value_vars=['det_blur'],)
df1
```

```python
X=df1.iloc[:,:2].values
X
```

```python
Y=df1.iloc[:,-1:].values
Y
```

```python
m,n=X.shape
b=0
w=np.zeros(n)
for i in range(10000):
    z=1/(1 + np.exp(-(X.dot(w)+b)))
    temp=(z-Y.T)
    temp=np.reshape(temp,m)
    dw=np.dot(X.T,temp)
    db=sum(temp)
    b=b-(db*0.001)
    w=w-(dw*0.001)
print("w=",w,"b=",b)
```

```python
z=X.dot(w)+b
```

```python
Y_Pred=1/(1 + np.exp(-z))#sigmoid function applied
```

```python
res=np.where( Y_Pred > 0.5, 1, 0 ) 
res
```

```python
correctly_classified = 0    


# counter    
count = 0    
for count in range( m ) :  

    if res[count] == Y[count] :            
        correctly_classified = correctly_classified + 1

    count = count + 1

print( "Accuracy on test set by our model       :  ", ( 
  correctly_classified / count ) * 100 )
```
